```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
# Merge the two DataFrames together based on the Dates they share

merge_table = pd.merge(dash_df, bitcoin_df, on="Date", how="inner")
merge_table.head()
```

```python
# Rename columns so that they are differentiated
#df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})

```

```python
# alternatively you can set your suffixes when the merge occurs
merge_table_df = pd.merge(dash_df, bitcoin_df, on="Date", how="inner", suffixes=("_Dash","_Bit"))
merge_table_df.head()
```

```python
# Collecting best open for Bitcoin and Dash
max_dash_open = merge_table_df["Open_Dash"].max()
max_bit_open = merge_table_df["Open_Bit"].max()

# Collecting best close for Bitcoin and Dash
max_dash_close = merge_table_df["Close_Dash"].max()
max_bit_close = merge_table_df["Close_Bit"].max()

# Collecting the total volume for Bitcoin and Dash
dash_vol = round(merge_table_df["Volume_Dash"].sum()/1000000,2)
bit_vol = round(merge_table_df["Volume_Bit"].sum()/1000000,2)
bit_vol
```

```python
# Creating a summary DataFrame using above values

summary_df = pd.DataFrame({"Best Bitcoin Open": [max_bit_open],
                           "Best Bitcoin Close": [max_bit_close],
                           "Total Bitcoin Volume": str(bit_vol) +" million",
                           "Best Dash Open": [max_dash_open],
                           "Besh Dash Close": [max_dash_close],
                           "Total Dash volume": str(dash_vol) +" million"})
```
